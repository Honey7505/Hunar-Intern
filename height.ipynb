```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("height1.csv")
print(df.columns)
```

```python
mean_he = df["Male's Height"].mean()
print("Mean_Height: ",mean_he,"cm")
```

```python
plt.figure(figsize=(8, 5))
sn.histplot(df["Male's Height"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Male Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
